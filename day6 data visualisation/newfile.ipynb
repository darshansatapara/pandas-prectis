```python
import matplotlib.pyplot as plt
views=[534,985,452,556,215,255,723,222]
days=range(1,9)
min_y=min(views)
x_co=days[views.index(min_y)]
plt.legend(loc="upper right")
plt.plot(days,views,label='you-tube view',color="r",marker="d",markerfacecolor="b",linestyle="--" ,linewidth="2")
plt.annotate("Lowest value",xy=(x_co,min_y),xytext=(1,215),arrowprops=dict(facecolor='black',shrink=0.01))
plt.grid(True,linewidth=1,linestyle="-.")
plt.xlabel("days")
plt.ylabel("daily views")
plt.title("you-tube daily views")
plt.show()
```
