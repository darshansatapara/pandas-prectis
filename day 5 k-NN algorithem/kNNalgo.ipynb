```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
diabetes.feature_names
```

```python
df = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df.head()
```

```python
df['target'] = diabetes.target
df.head()
```

```python

df[df.target==151].head()
```

```python
df[df.target==75].head()
```

```python
df[df.target==109].head()
```

```python
# Create a new column based on a transformation of the target variable
df['new_diabetes'] = df['target'].apply(lambda x: 'High' if x > df['target'].mean() else 'Low')
print(df.head())
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
feature1 = 'bmi'  # Replace with the desired feature name
```

```python
feature2 = 'bp'
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df[feature1], df[feature2], c=df['target'], cmap='viridis', marker='o')
plt.colorbar(label='Target Variable')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title(f'Scatter Plot: {feature1} vs {feature2} colored by Target Variable')
plt.show()
```
